```python
import os,sys
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.tri as tri     # for making triangular contours
import numpy as np
import array
import math
from operator import truediv
from operator import add
import pylab as plt
```

```python
# Define empty lists and constant parameters

beta = []
alpha = []
sinba = []
TanB = []
M12 = []
Mh = []
MH = []
MA = []
Cpl3 = []
vev_sqr = pow(246,2)


# Define path to scanned points directory

directory ='ValidPoints'
spcfiles =os.listdir(directory)
print(len(spcfiles))

# Loop overall spc files

for filename in spcfiles:
    spcfile_path = os.path.join(directory,filename)        # source spc file to be renamed
    if os.path.isfile(spcfile_path):
        #print('Found file '+spcfile_path)
        with open(spcfile_path,'r') as txt_spcfile:
            lines = txt_spcfile.readlines()
            for line in lines:
                if line[26:31] == 'betaH':
                    beta_inRad = float(line[8:18])*0.1
                    beta_inDeg = math.degrees(beta_inRad)
                    beta.append(beta_inDeg)
                    
                if line[26:32] == 'alphaH':
                    alpha_inRad = float(line[8:18])*0.1
                    alpha_inDeg = math.degrees(alpha_inRad)
                    alpha.append(alpha_inDeg)
                    
                if line[27:34] == 'TanBeta':
                    tanB = float(line[9:19])*0.1
                    TanB.append(tanB)
                    
                if line[33:37] == 'hh_1':
                    mh = float(line[15:29])
                    Mh.append(mh)
                    
                if line[33:37] == 'hh_2':
                    mH = float(line[15:29])
                    MH.append(mH)
                    
                if line[33:37] == 'Ah_2':
                    mA = float(line[15:29])
                    MA.append(mA)
                    
                if line[27:35] == 'M12input':
                    m12 = abs(float(line[8:23]))
                    M12.append(m12)
                    
                if line[27:39] == 'Lambda3Input':
                    lamda3 = float(line[9:23])
                    Cpl3.append(lamda3)
                     
                
    else:
        print('File '+spcfile_path+' not found!')
        


for item1, item2 in zip(beta,alpha):
    item = item1 - item2
    sin_b_a = math.sin(item)
    sinba.append(sin_b_a)
    
    
# Calc items for contour plots

item11 = []     # stores v^2 * lambda3 term values
item12 = []     # stores 2 MA^2 values
item13 = []     # stores sin2B values
item14 = []     # stores MH^2 - Mh^2 values
item15 = []     # stores sin2B/(MH^2 - Mh^2) values
item16 = []     # stores (v^2 * lambda3) - (2 MA^2)  values
item17 = []     # stores the value of all terms = sin2B
#m12_sqr = np.nan  # nan -> not a number variable stores nothing


for l3 in Cpl3:
    item_ = vev_sqr*l3
    item11.append(item_)
    
for m in MA:
    item2_ = 2*pow(m,2)
    item12.append(item2_)
    
for b in beta:               # angle is in radians
    item3_ = math.sin(2*b)
    item13.append(item3_)
    
for h,H in zip(Mh,MH):
    item4_ = pow(H,2)-pow(h,2)
    item14.append(item4_)
    
for i,j in zip(item13,item14):
    item5_ = i/j
    item15.append(item5_)
    
for k,l in zip(item11,item12):
    item6_ = k-l
    item16.append(item6_)
    
'''
for i1,i2,i3 in zip(item16,item13,item15):
    item7_ = (i1 + ((2*m12_sqr)/i2))*i3
    item17.append(item7_)
'''

#inverse_sin = math.arcsin()

list1_ = list(np.array(M12)*2)
list2_ = list(map(truediv,list1_,item13))
list3_ = list(map(add,item16,list2_))
list4_ = list(map(lambda a,b:a*b,list3_,item15))
list5_ = list(map(lambda d: d/2,list4_))
list55_ = np.array(list5_)
list_ = np.array(list(map(np.float_, list55_)))
list__ = np.array(list(map(np.float_, TanB)))

    
    
    
#print('item17 list size is ' + str(len(item17)) + ' with element item17[200] = ' + str(item17[200]))


# create 2D contours 

x,y = np.meshgrid(M12,TanB)
#z = math.sin((np.arctan(list__))-(np.arcsin(list_)))


    
############################################
# Create plots

colors = np.random.rand(363)     # define an array of colors randomly

plt.scatter(Mh,TanB,color='m',marker='.')
plt.scatter(Mh, TanB, c=colors, cmap='plasma', marker='.')
plt.xlabel(r'$M_{h} [GeV]$')
plt.ylabel(r'$\tan \beta$')
plt.colorbar()
plt.show()

plt.scatter(Mh,M12,color='m',marker='.')
plt.xlabel(r'$M_{h} [GeV]$')
plt.ylabel(r'$m_{12}^{2}$')
plt.show()
    
plt.tricontourf(M12,TanB,Mh, cmap='plasma')
#plt.colorbar()
cbar = plt.colorbar()
cbar.set_label(r'$M_{h} [GeV]$')
plt.xlabel(r'$m_{12}^{2}$')
plt.ylabel(r'$\tan \beta$')
#plt.zlabel(r'$M_{h} [GeVs]$')
plt.xlim(500,3000)
plt.ylim(0.25,0.6)
plt.show()

# Interpolation from triangular grid into quad grid contour

xi, yi = np.meshgrid(np.linspace(0, 3, 20), np.linspace(0, 3, 20))   # Interpolate to regularly-spaced quad grid.

interp_cubic_geom = tri.CubicTriInterpolator(x, y, Mh, kind='geom')      # x = M12, y = TanB, z = Mh
zi_cubic_geom = interp_cubic_geom(xi, yi)
    
    
##############################################

i = 0

while(i < len(sinba)):
    #print('P' +str(i)+'--->' + ' '+str(sinba[i]))
    if sinba[i] < 0.01 and sinba[i] > -0.01:
        print('P' +str(i)+'--->' + ' '+str(sinba[i]))
    i = i+1
    
'''    
print('P18---> ' + str(TanB[18])) 
print('P127---> ' + str(TanB[127])) 
print('P263---> ' + str(TanB[263])) 
print('P299---> ' + str(TanB[299])) 
print('P352---> ' + str(TanB[352])) 

print('P18---> ' + str(Mh[18])) 
print('P127---> ' + str(Mh[127])) 
print('P263---> ' + str(Mh[263])) 
print('P299---> ' + str(Mh[299])) 
print('P352---> ' + str(Mh[352]))

print('P18---> ' + str(MH[18])) 
print('P127---> ' + str(MH[127])) 
print('P263---> ' + str(MH[263])) 
print('P299---> ' + str(MH[299])) 
print('P352---> ' + str(MH[352]))

print('P18---> ' + str(MA[18])) 
print('P127---> ' + str(MA[127])) 
print('P263---> ' + str(MA[263])) 
print('P299---> ' + str(MA[299])) 
print('P352---> ' + str(MA[352]))

print('P18---> ' + str(M12[18])) 
print('P127---> ' + str(M12[127])) 
print('P263---> ' + str(M12[263])) 
print('P299---> ' + str(M12[299])) 
print('P352---> ' + str(M12[352]))

   
print("¨¨¨tan B values¨¨¨")

for j in TanB:
    print(j)
    
print("¨¨¨Mh values¨¨¨")
    
for k in Mh:
    print(k)
    
for i,j,k in zip(sinba,TanB,Mh):
    print (str(i) +','+ str(j) + ','+ str(k))
'''    
        
print(len(sinba))      
print(len(alpha))
print(len(beta))
print(len(TanB))
print(len(Mh))
print(Mh[200])
print(len(MH))
print(len(MA))
print(len(M12))
print(len(Cpl3))
```
